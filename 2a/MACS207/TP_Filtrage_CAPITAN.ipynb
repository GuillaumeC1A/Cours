<h3> <strong> Étape 1 </strong> - Importations, initialisations, rappel du problème et plan </h3>

Je vais commencer par brièvement rappeler le problème. Nous étudions un changement de régime dans un processus de Poisson à un instant donné $\tau$ qui suit une loi exponentiellelede paramètre $\mu = 0.2$.
</br>
L'idée est qu'avant $\tau$ on observe un processus de Poisson $N_a$ d'intensité $a$, indépendant du processus de Poisson $N_b$ d'intensité $b$ observé après $\tau$. C'est une autre manière de voir un processus qui saute d'intensité.
</br>
Le but de notre travail est d'estimer $\tau$ en se penchant sur les instants de saut observés. Ils n'auront pas la même densité, la même <strong> intensité </strong> avant et après $\tau$. Pour ce faire, on va calculer la fonction de filtrage, résoudre l'équation différentielle qui nous donne $\hat{Z}$ et enfin trouver $\tau$ à partir de ça. On comparera enfin cette méthode à celle du K-means, qui "regroupe" les sauts par paquets, et on cherchera $\tau$ entre les deux paquets trouvés. 
</br> Quelques importations pour commencer :

<h3> <strong> Étape 2 </strong> - Fonction de filtrage h </h3>

On a $h(s) = \dfrac{f(s)}{1-F(s)}$

Comme $\tau$ suit une exponentielle de paramètre $\mu$ on a comme densité $f(s) = \mu e^{-\mu x}$ et comme fonction de répartition $F(s) = 1 - e^{-\mu x}$ ce qui donne :
$$h(s) = \dfrac{\mu e^{-\mu x}}{1-(1 - e^{-\mu x})} = \dfrac{\mu e^{-\mu x}}{e^{-\mu x}} = \mu$$

On tombe sur une constante, ce qui nous arrange bien car cela évitera les erreurs numériques.

<h3> <strong> Étape 3 </strong> - Trajectoire de N sur [0,10] </h3>

On veut donc modéliser notre processus de Poisson qui part d'abord en 0 d'une intensité $a$ puis qui arrive en 10 avec une intensité $b$, ayant changé d'intensité en cours de route.

```python
def poissonAB(n, a=a, b=b, tau = tau, T0 = 0, T1 = 10):
    jumps = [0]
    draw_count = 0
    t = T0 #Ici, 0 jusqu'à 10
    
    #0<t<tau - intensité a
    while t<min(tau,T1): #J'ai parfois des tau supérieurs à T. Le TP n'est alors pas très intéressant...

        draw = np.random.exponential(scale = 1/a) #on tire d'abord des poisson d'intensité a
        t += draw
        draw_count += 1
        jumps.append(t)
        
    #On mémorise le changement de régime :
    chang_reg = draw_count
    
    #0<t<T - intensité b
    while t<T1: #si tau était plus grand que T1, on ne ferait rien ici.
        
        draw = np.random.exponential(scale = 1/b)
        t += draw
        draw_count += 1
        jumps.append(t)
    
    return jumps, chang_reg
```

<h3> <strong> Étape 4 </strong> - Représentation de $\hat Z$ et estimation de $\tau$ </h3>

On s'intéresse donc, comme suggéré par l'exercice du polycopié, à $\hat Z$ (que l'on notera Z_hat dans notre code). Le processus est défini par :
$$t \mapsto  \hat Z(t) = \mathbb{E}[Z(t)|\mathcal{F}_t^N] $$ avec $\mathcal{F}_t^N$ la filtration naturelle du processus N. 
</br>
Dans l'exercice on montre que (*Q9*) $\forall t \in [T_n, T_{n+1}]$,
$$\hat Z(t) = \hat Z(T_n) + \int_{T_n}^{t} (h(s)-(b-a)\hat Z(s))(1-\hat Z(s)) \mathrm{d}s ~~~~~~~~~(R1)$$
Et on a également la formule pour calculer l'amplitude du saut (*Q10*):
$$\hat Z(T_n) - \hat Z(T_n^{-}) = \dfrac{(b-a)\hat Z(T_n^{-})(1-\hat Z(T_n^{-}))}{a+(b-a)\hat Z(T_n^{-})} ~~~~~~~~~(R2)$$

Cette pseudo-discrétisation sur les sauts nous permet sur chaque intervalle $[T_n, T_{n+1}]$ d'établir l'équa diff suivante pour $y = \hat Z$:

$$
\left\{
    \begin{array}{ll}
        y'(t) = f(t,y(t)) = (h(t)-(b-a)y(t))(1-y(t)) = (\mu-(b-a)y)(1-y)\\
        y(T_{n-1}) = \hat Z(T_{n-1})
    \end{array}
\right.
$$

```python
#On définit d'abord la fonction de l'équation différentielle f(y,t) pour la rentrer plus tard dans le solver odeint de scipy.integrate:
def f(y,t):
    return (mu-(b-a)*y)*(1-y)
```

```python
def R2_func(a,b,Tn_m,Z_hat):
    return ((b-a)*Z_hat[Tn_m]*(1-Z_hat[Tn_m])) / (a + (b-a)*(Z_hat[Tn_m]))
```

```python
def Z_hat_f(n, process = poisson1, a=5, b=10):
    
    intervalle_cur = np.linspace(process[0],process[1],n, endpoint=False) #résolution de l'équa diff sur cet intervalle courant
    #et comme on va les concaténer pour fabriquer notre abscisse... on veut un crochet ouvert à droite
    intervalle_gen = intervalle_cur #notre abscisse générale qui serait à la fin la concaténation de tous les intervalles de résolution.
    
    Z_hat_gen = odeint(f,process[0],intervalle_cur).reshape(-1) #la résolution en elle-même, sur ce premier intervalle
    Z_hat_jumps = [0] #Ajouté a posteriori pour la recherche de Tau, avoir la valeur aux instants de saut est indispensable
    
    for i in range(1,len(process)-1):
        Z_jump = Z_hat_gen[-1] + R2_func(a,b,-1,Z_hat_gen) #En fait pour avoir Tn_m = Tn_moins il suffit juste de prendre toujours le dernier élément de la liste. Donc [-1]
        Z_hat_jumps.append(Z_jump)
        
        intervalle_cur = np.linspace(process[i], process[i+1],n, endpoint=False) #actualisation du nouvel intervalle courant
        intervalle_gen = np.concatenate((intervalle_gen,intervalle_cur))
        
        Z_hat_cur = odeint(f,Z_jump,intervalle_cur).reshape(-1) #résolution sur le nouvel intervalle
        Z_hat_gen = np.concatenate((Z_hat_gen,Z_hat_cur)) #Ajout de la nouvelle résolution à la générale
        
    Z_hat_jumps.append(Z_hat_gen[-1]) #on ajoute la valeur finale à la liste
    return intervalle_gen, Z_hat_gen, Z_hat_jumps
```

```python
absc_t, Z_hat, Z_hat_jumps = Z_hat_f(1000)
```

```python
plt.close()
plt.figure(figsize=(12,7))

plt.plot(absc_t, Z_hat)
plt.title('Z_hat')
plt.axvline(x=tau, color = 'brown', label = 'Vrai Tau')
plt.xlabel('temps t')
plt.legend()
plt.show()
```

On constate bien une montée très rapide juste après $\tau$ pour $\hat Z$. Ma méthode (la première que je vais implémenter pour l'instant en tout cas) va être de chercher maintenant à voir à quel moment précisément $\hat Z$ dépasse le seuil 0.5. En inversant la fonction on retrouvera le bon $t=\tau$ (i.e. informatiquement à quel indice on y arrive)

```python
def tau_search(Z_hat = Z_hat,tau_sei = a/b):
    for i in range(len(Z_hat)):
        if Z_hat[i]>tau_sei:
            return i
        #return fonctionne comme un break, dès qu'on trouve un t qui fonctionne on le renvoie et on s'arrête du même coup.
    return -1 #si ça n'a pas trouvé, par exemple s'il n'y a pas eu de changement de régime où qu'il était trop ténu. 
```

```python
tau_est_absc = tau_search() #Ceci renvoie donc l'indice du tau, et pour retrouver tau on va au bon indice dans la liste des t.
tau_est = absc_t[tau_est_absc]
```

```python
print('Notre tau estimé est '+str(tau_est)+' contre un tau réel de '+str(tau)+' soit une erreur relative de '+str(abs((tau_est-tau)/tau)))
```

Cette méthode nous donne une erreur relative de 5% ce qui est excellent ! (En réalité j'ai pris ici un cas où mon code marchait plutôt bien, en général on est plutôt vers 10-15% mais cela reste très très bien).

<h3> <strong> Étape 5 </strong> - K-means, deuxième estimation de $\tau$ </h3>

La difficulté principale à laquelle on se heurte pour faire une méthode k-means est la donnée sur laquelle on doit appliquer notre méthode. Quelle caractéristique souhaite-t-on dégager, qui est suffisamment marquante pour qu'on distingue bien deux groupes (ce que fait la méthode en pratique) autour de $\tau$ ? J'ai opté pour quelque chose qui me paraissait pertinent : ce que j'ai tracé plus haut sur les graphes, la pseudo-dérivée, l'intensité, l'accroissement moyen, bref autant de noms pour parler de ce qui tend vers $a$ sur la partie du processus $N_a$ et vers b sur la partie du processus $N_b$, la quantité $\dfrac{N(t)}{t}$
</br>
C'est à dire qu'à partir de la trajectoire que l'on a on va créer de la donnée en $\dfrac{N(T_{n})}{T_n}$ et on va lui assigner un label dépendant de sa valeur. Si elle est assez haute, la donnée en question sera dans le groupe $b$. Sinon, groupe $a$. En moyenne, lorsque l'accroissement dépasse $\dfrac{a+b}{2}$ on devrait être dans la partie 2 du processus. 
</br> 
A noter qu'à l'origine j'avais bien pris comme donnée $\dfrac{N(T_{n})}{T_n}$, mais à cause de problèmes de calibration le k-means inversait les 0 et les 1 pour la classification. Ce qui ne posait pas de problème fondamentalement, il arrivait toujours à identifier deux tendances, deux clusters, mais juste à l'opposé de ce qu'il fallait. Pour remonter à $\tau$ de manière automatique, il me fallait quelque chose d'exact : c'est là que j'ai eu l'idée de recalibrer N(t) par sa tendance bT[i] pour obtenir N(t)-bT[i] (qui correspondrait à la moyenne en cet endroit par rapport à l'instant de changement de régime)
</br>


```python
def k_means_data1(process = poisson1, a=5, b=10):
    data = []
    for i in range(2,len(process)-1): #On enlève l'initialisation
        data.append((i-b*process[i])/process[i]) #"nouvelle version" du N(t)/t instantané. En d'autres termes, c'est une intensité instantanée re-calibrée.
        #1/process[i] devrait être proche de l'intensité a ou b selon qu'on est avant tau ou après. 
    return data[:-1]
```

```python
#Créons notre liste de données pour le kmeans. 
data1 = k_means_data1()
plt.plot(range(len(data1)), data1)
```

On constate qu'en utilisant les accroissements, l'intensité instantanée (qu'ils soient recalibrés ou non, encore une fois c'est juste une affaire de label), à part sur ces deux pics assez étranges au départ, on a bien une tendance exploitable. Voyons les résultats avec un kmeans :

Pour trouver ensuite notre fameux $\tau$, on va d'abord remonter eux deux clusters différents (correspondant aux deux régimes) et ensuite revenir à $\tau$ en faisant une moyenne des extrémités des deux clusters, càd la moyenne entre le dernier instant du premier cluster et le premier instant du deuxième.

(à nouveau, ici il n'y a pas de pics parasites... c'est le problème des maths aléatoires, c'est la reproductibilité...) </br>
La plupart des formules que je voulais implémenter (notamment le basique (processus[zeros[-1]] + processus[ones[0]])/2 ne fonctionne pas pour la bonne et simple raison qu'il y a des 1 parasites. Et donc que le $\tau$ est identifié bien trop tôt (sans mauvais jeu de mot) à tord. Pour résoudre le problème, je vois deux solutions :
</br> <strong>Lisser l'ensemble </strong>en superposant une sorte de nouveau kmeans (je prends la valeur majoritaire sur les voisins immédiats) afin d'<strong>accentuer encore plus les tendances déjà dégagées</strong>. </br>
<strong> Utiliser une nouvelle formule qui prendrait en compte non pas seulement les extrémités mais l'ensemble des sauts / instants attribués à un groupe en particulier. </strong> 


```python
#Implémentation de la solution 1: (devient obsolète lorsque le processus est aussi lisse et proche de la tendance que celui que j'ai, mais utile en général)
def lissage_N(L,N=3):
    M = np.zeros(len(L))
    M[0] = 0 #Eviter les problèmes de bord
    M[-1] = 1 #idem
    outed = int((N-1)/2)
    for i in range(outed,len(M)-outed): #Pas les extrémités
        M[i] = round((L[i-1]+L[i]+L[i+1])/3)
    return M
```

```python
#On obtient bien deux composantes connexes, de zéro et de un de chaque côté. Calculons maintenant avec notre formule le Tau du kmeans :
# "Nouvel" ensemble des 0 prédits, i.e. du premier cluster trouvé par le k-means.
zeros = np.where(labels_liss==0)[0] #dernier élément correspondant à Na
# "Nouvel" ensemble des 1 prédits, i.e. du deuxième cluster trouvé par le k-means.
ones = np.where(labels_liss==1)[0] #Premier élément correspondant à Nb. 
tau_kmeans = (poisson1[zeros[-1]] + poisson1[ones[0]])/2 #Moyenne entre le dernier saut classé pour Na et le premier saut classé pour Nb
print("On trouve comme Tau du kmeans : "+str(tau_kmeans))
print("En comparant avec le vrai Tau = "+str(tau)+", on trouve une erreur relative de "+str((abs(tau_kmeans-tau))/tau_kmeans))
```
