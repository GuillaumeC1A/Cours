Implémentez la méthode des $k$-plus proche voisins en prenant comme caractéristiques les valeurs des
pixels. On donnera, pour différentes valeurs du paramètre k (k=1, 3, 4, 5):
- Le taux d’erreur.
- La matrice de confusion - La matrice de confusion est une matrice de taille $10\times10$ dont l’élément d’indice (i,j) est le nombre de vecteurs de la classe i qui ont été affectés à la classe j par la méthode.
- Visualisez quelques confusions entre caractères.

```python
def imDist(A,B):
    err = 0
    for i in range(len(A)):
        err += (A[i] - B[i])**2
    return np.sqrt(err)
#norme euclidienne dans la distance des pixels
```

```python
def removeZeroi(A,i):
    return A.pop(A.index((0.0,i)))
```

```python
def mostLittle(A):
    mini = A[0][1]
    ind = 0
    for i in range(len(A)):
        if A[i][0] < mini and A[i] > 0:
            mini = A[i]
            ind = i
    return mini,i
```

```python
def maxNow(A):
    maxi = A[0][0]
    ind = 0
    for i in range(len(A)):
        if A[i][0] > maxi:
            maxi = A[i][0]
            ind = i
    return maxi,ind
```

```python
def k_ppv(train_data, train_labels, k):
    L = len(train_data)
    k+=1
    score=0
    for i in range(L):
        datai = train_data[i]
        distList=[]
        for j in range(k):
            distList.append((imDist(datai,train_data[j]),j)) #On dispose maintenant de k distances que l'on va actualiser à chaque fois qu'un voisin plus proche est traité. 
        curmax = maxNow(distList)
        lastMaxInd = curmax[1]
        for j in range(k, L):                
            dataj = train_data[j]
            distij = imDist(datai,dataj)
            if (distij < curmax[0]):
                distList[curmax[1]] = (distij,j)
                curmax = maxNow(distList) 

        removeZeroi(distList,i)
        
    #On dispose maintenant des k plus proches voisins pour i. A la différence des autres codes que vous pourrez lire, je ne fais pas d'énorme comparaison de k plus proches distances 
    #sur 1000 éléments, j'ai privilégié une solution plus élégante qui consiste à ne garder toujours que les k plus petites distances, et à remplacer la plus grande de celles-ci
    #par une plus petite quand on en rencontre une. Cela demandait un peu de finesse sur la façon de coder (garder les indices en mémoire), le pire est que c'est moins efficace.
    #La lenteur du programme est énorme, le seul avantage est qu'on utilise une place mémoire très très faible (normal, la relation est inverse). Temps moyen : 11 minutes k=2
        
        labelsFound = []
        for j in range(len(distList)):
            labelsFound.append(train_labels[distList[j][1]])
        A = [0]*11
        
        for j in range(1,11):
            A[j] = labelsFound.count(j) #le range(1,11) est pour avoir le 10 correspondant au 0.
        pred.append(A.index(max(A)))

        if A.index(max(A)) == train_labels[i]: #S'il y a deux maximums, on prend le plus petit indice, une décision arbitraire devant une situation de toute façon non couverte.
            score += 1
            
        if i%100 == 0:    
            print('Tour '+str(i)+'. Les '+str(k-1)+' plus proches sont :')
            print(distList)    
            print('Notre score est de '+str(score))
    return score, pred #On retourne aussi les prédictions pour pouvoir construire la matrice de confusion plus tard 
```

- Construisez des prototypes (au moins un par classe) à partir des images de l’ensemble d’apprentissage, et visualisez-les. Le prototype peut être par exemple la moyenne des exemples correspondants à chaque classe. 
- Classez les images de test en les comparant aux prototypes.
- Comparez avec les résultats obtenus précedemment.

```python
def prototype(classe):
    proto = np.zeros(np.shape(train_data[0]))
    norm = 0
    for i in range(len(train_data)):
        if train_labels[i] == classe:
            proto +=train_data[i]
            norm +=1
    return proto / norm
```

```python
#Tous les chiffres mal classés pour k=2 : 
f, ax = plt.subplots(2,5) 
f.set_size_inches(12, 5.5)
i,j = 0,0
for l in range(1,11):
    ax[i,j].imshow(prototype(l).reshape((28,28)), cmap = 'gray')
    ax[i,j].set_title('Prototype pour '+str(l))
    j+=1
    if j==5:
        j=0
        i+=1
```

Ces chiffres sont cependant très reconnaissables ! Les 2 ont bien leur boucle, le trois à bien ses deux courbes, le 4 est bien génométrique, le 0 bien rond, le 8 a bien ses deux boucles, le 6 la sienne... Le 1 est bien vertical également, et le 5 ressemble à un 5 parfait : géométrique en haut et bien courbé au centre en bas.

```python
def classify_proto(test_data, test_labels, prototypes):
    score=0
    pred = []
    for i in range(len(test_data)):
        distList = []
        for j in range(10):
            distList.append(imDist(test_data[i], prototypes[j]))
        bestDist = np.argsort(distList)[0] #On ne garde que les k indices les plus bas pour obtenir les k premières distances. On élimine la plus basse qui vaudra toujours 0 (dist(i,i)).
        bestDist+=1 #A cause du décalage dans les labels de prototype dû au 0 <-> 10 : le 1 est à l'emplacement 0.
        if bestDist == test_labels[i]: #S'il y a deux maximums, on prend le plus petit indice, une décision arbitraire devant une situation de toute façon non couverte.
            score += 1
        pred.append(bestDist)
        if i%100 == 0:    
            print('Tour '+str(i)+'. Le prototype le plus proche est:')
            print(bestDist)    
            print('Et ce pour une image de label '+str(test_labels[i])+'.')
            print('Notre score est de '+str(score))
    return score, pred
```

```python
score, pred = classify_proto(test_data, test_labels, [prototype(i) for i in range(1,11)])
print('Score final :'+str(score))
print('Sur un nombre total de '+str(len(test_data)))
print('Ratio : '+str(221/300)+'à comparer avec 0.865')
#print(pred)
#print(test_labels)
```
