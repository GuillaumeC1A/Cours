## I - Metropolis-Hastings vs Rejection sampling
We would like to sample from the density function:
$$ f(x) \propto  x^4\exp(- x^2) $$ 
Using the proposal: $g(x) \propto \exp(- \frac{1}{2} x^2) $.

1. Find an empirical bound M so as to apply rejection sampling using g to sample from f and visualize the domination of f by Mg.

```python
x = np.linspace(-10, 10, 1000)
f = lambda x: np.exp(- x ** 2) * x ** 4
g = lambda x: np.exp(- 0.5 * x ** 2)
```

```python
#In order to check the minimal constant so that Mg > f everywhere :
M,m = 1,1
for val in x:
    m= f(val)/g(val)
    if m>M:
        M=m
print(M)
```

We found the bound above, which we will round up to 2.17 (the closer Mg is to f, the more likely we are to accept a given generated sample). Let's visualize how it goes :

```python
M=2.17
plt.close('all')
plt.plot(x, f(x), label='f dominated curve')
plt.plot(x,M*g(x), label='Mg dominate curve')
plt.title('Domination of f by Mg')
plt.legend()
plt.show()
```

2. Implement the rejection sampling algorithm and visualize the empirical histogram against the density f. Use scipy.integration to estimate the normalizing constant of f for accurate visualization of the density.

```python
#A few useful things first :
import time
import scipy.integrate as integrate
from scipy.stats import norm

def iCDF_g(x): #inverse cumulative distribution function (iCDF) of g. Here it is simple, but for "good practice", I isolate it.
    """Application of the inversion theorem"""
    #ppf is the inverse of the gaussian CDF.
    return norm.ppf(x) 

#I'll just add a comment here : you may have noticed I did NOT re-normalize g or the CDF before taking the iCDF. 
#This is because we re-normalized g anyways in the first question, by taking M = 2.17
#If we chose instead g* = 1/(2pi) exp(-x²), the M would have been different... The M we took corresponds to a non-factored g !
    
norm_cons_f = integrate.quad(f,-np.inf,np.inf)[0]
print('The normalizing constant for f is roughly '+str(norm_cons_f))
```

```python
def rejection_sampling(n=1000, f=f, g=g, M=M, N=10000, bins=50):
    """Reject sampling algorithm for f with Gaussian proposal g.
    n number of samples generated, not accepted !
    N number of required samples for the algorithm to finish"""
    
    start_t = time.time()
    samps = []
    while len(samps)<N:
        
        #Sample generation of n (y,u) pairs.
        y = iCDF_g(np.random.uniform(0,1,n))
        u = np.random.uniform(0,1,n)
        
        #Accepting condition - I encountered the np.where function in a previous work. It's great !
        indices = np.where(u<=f(y)/(M*g(y)))
        
        #This is where you see I did not need to put a np.sqrt(2*np.pi) in the iCDF.
        #y contains samples from a true gaussian, yes, but anyways we only accept those who satisfy our condition.
        #and that condition includes a re-normalization anyways, in the M ! It's just a 'value', a 'size' difference, that gets corrected through the test.
        #Once again, if we chose g as a true gaussian, with the factor in front, our M would have been different, thus the test as well, and we would have had the same result.
        
        #I now have all the indices of accepted samples. I'm extending the samples list with the new accepted ones.
        samps.extend(y[indices])
    end_t = time.time()
    #As soon as we have our samples, let's show our result :
    plt.close('all')
    plt.hist(samps, density=1, bins = bins, label='rejection sampling method')
    plt.plot(x,f(x)/norm_cons_f,label='goal : density of f', linestyle='--')
    plt.title('rejection sampling algorithm, N = '+str(N)+', bins='+str(bins))
    plt.legend()

    print('Elapsed time of rejection sampling method : '+str(end_t-start_t)+' seconds')
    return plt.show()
```

3. Implement the random walk metropolis hastings algorithm with a Gaussian conditional probability and visualize the empirical histogram against the normalized density f/F

What we first did is to sample from f using g, even though they are not the same at all ! g dominates f so it's good for rejection sampling, but it doesn't work at all for a metropolis algo...
So what we do now with the Metropolis-Hastings is we use discrete values of f (which we can compute) to get back to the density (harder to sample from). 

```python
def rw_metropolis_hastings(n=10000, f=f, bins=40, seed=None):
    """Random walk metropolis hastings algorithm for f with Gaussian proposal g. But g is not in arguments so I did without.
    n: int.
        number of samples.
    f: callable.
        density we want to sample from.
    seed: int (optional).
        random seed initialization.
    """
    start_t = time.time()
    np.random.seed(seed)
    xt = np.random.uniform(-1,1) #x0
    samps = [xt]
    
    #As mentioned in the question, we will use the gaussian distribution for the 'transition' probability. (so going left is as probable as going right, is it symmetric)    
    f_x = f(xt)
    while len(samps) < n:
        x_ = xt + np.random.normal()   #Gaussian centered on x. I could have also implemented it as x_ = g(x, 1) with x the mean and 1 the std deviation. But it's fine like this.
        f_x_ = f(x_)
        alpha = f_x_/f_x
        if alpha>= 1:
            #if the ratio between g(x_) and g(x) is already higher than 1, we necessarily move & update.
            xt = x_
            f_x = f_x_
        #if not, we draw a sample from Uniform[0,1] and if it is smaller than our alpha ratio, we keep it.
        uni_ = np.random.uniform(0,1)
        if uni_ <= alpha:
            xt = x_ 
            f_x = f_x_
        #else: #In case of failure
            #x = x stay in position, not necessary
            #f_x = f_x stay in position, not necessary
        samps.append(xt)
        
    end_t = time.time()
    #As soon as we have our samples, let's show our result :
    plt.close('all')
    plt.hist(samps, density=1, bins = bins, label='Metropolis Hastings method')
    plt.plot(x,f(x)/norm_cons_f,label='goal : density of f', linestyle='--')
    plt.title('Metropolis Hastings algorithm, n = '+str(n)+', bins='+str(bins))
    plt.legend()

    print('Elapsed time of Metropolis Hastings method : '+str(end_t-start_t)+' seconds')
    return plt.show()
```

# II - Gibbs sampling
Gibbs sampling is a Markov Chain sampling algorithm used to sample from a multivariate random variable by sampling each component from its conditional distribution while the others components are fixed.

We would like to sample from a multivariate Gaussian $(X_1, X_2)$ with density:
$$p(x_1, x_2) = \frac{1}{2\pi\sqrt{1 - \rho^2}} \exp\left(- \frac{1}{2(1 - \rho^2)}(x_1^2 + x_2^2 - 2\rho x_1 x_2)\right) $$
*1. Identify the parameters of the distribution and compute the conditional distributions of each component.*

First, for a reminder, for a dimension 2 multivariate density, we have 
$$p(x_1, x_2) = \dfrac{1}{2 \pi \sqrt{|det(\Sigma)|}} \exp(-\dfrac{1}{2} (x-\mu)^T \Sigma^{-1} (x-\mu)) $$

After a few trials and eliminations, for the symmetric case we have ($X_1$ and $X_2$ play the same role) we get :
</br> Covariance matrix
$$\Sigma = \begin{bmatrix}
1 & \rho\\
\rho & 1
\end{bmatrix}$$ 
with its inverse 
$$\Sigma^{-1} = \dfrac{1}{1-\rho^2}\begin{bmatrix}
1 & -\rho\\
-\rho & 1
\end{bmatrix}$$
Mean vector 
$$\mu = \begin{bmatrix}
0\\
0 
\end{bmatrix}$$
Which we can rewrite as $$(X_1,X_2) \sim \mathcal{N}\begin{pmatrix}
\begin{pmatrix}
0\\
0
\end{pmatrix} ,  \begin{pmatrix}
1 & \rho\\
\rho & 1
\end{pmatrix}\\
\end{pmatrix}.$$

Now for the conditional distributions, let's compute $p(x_1|x_2)$, as the problem is symmetric it will give us exactly the other one by switching indices :
We have first the marginal densities ("plain old" normal distributions) : $$p(x_i) = \dfrac{1}{\sqrt{2\pi}}exp{(-\dfrac{x_i^2}{2})}$$
</br> 
Each conditional density is given by the Bayes formula:
$$ p(x_1|x_2) = \dfrac{p(x_1,x_2)}{p(x_2)} = \dfrac{\frac{1}{2\pi\sqrt{1 - \rho^2}} \exp\left(- \frac{1}{2(1 - \rho^2)}(x_1^2 + x_2^2 - 2\rho x_1 x_2)\right)}{\dfrac{1}{\sqrt{2\pi}}exp{(-\dfrac{x_2^2}{2})}} = 
\dfrac{1}{\sqrt{2\pi(1-\rho ^2)}} \exp\left(-\dfrac{x_1^2+x_2^2-2\rho x_1x_2+(-x_2^2+x_2^2 \rho^2)}{2(1-\rho^2)}\right)=
\dfrac{1}{\sqrt{2\pi(1-\rho ^2)}} \exp\left(-\dfrac{x_1^2-2\rho x_1x_2+x_2^2 \rho^2}{2(1-\rho^2)}\right)$$
i.e. 
$$p(x_1|x_2) = \dfrac{1}{\sqrt{2\pi(1-\rho ^2)}} \exp\left(-\dfrac{(x_1-\rho x_2)^2)}{2(1-\rho^2)}\right)$$ 
Which can be rewritten as 
$$X_1 | (X_2=x_2) \sim \mathcal{N}(\rho x_2, 1-\rho^2)$$
by identification. 
</br> Symmetrically, we have 
$$p(x_2|x_1) = \dfrac{1}{\sqrt{2\pi(1-\rho ^2)}} \exp\left(-\dfrac{(x_2-\rho x_1)^2)}{2(1-\rho^2)}\right)$$ 
and 
$$X_2 | (X_1=x_1) \sim \mathcal{N}(\rho x_1, 1-\rho^2)$$

*2. Implement the associated Gibbs sampler where each $X_1, X_2$ are sampled sequentially from the conditional distributions.*

```python
def gibbs_sampler(n=1000, rho = 0.7):
    """Gibbs sampler for (X_1, X_2)."""
    X = []
    x1,x2 = 0,0 #Origin of the samples, center of the target distribution (we don't want to start far from it... but we'll get back to that later)
    for i in range(int(n)):
        #we actualize x1 first, by convention, but as the situation is symmetric it doesn't change anything. 
        x1 = np.random.normal(rho*x2, 1-rho**2)
        x2 = np.random.normal(rho*x1, 1-rho**2)
        
        #We append the tuple generated from their respective conditional distributions
        X.append([x1,x2])
    return X
```

4. Using the numpy corrcoef function, plot the auto-correlation values as a function of the lag: $corr(X_t, X_{t - lag})$

```python
#I will compute both autocorrelations (I mean corr(X_1t, X_1t-lag) AND corr(X_2t, X_2t-lag), but as the situation is completely symmetrical, we of course expect the same results. 
#Given the construction of our sample generation function (gibbs_sampler), which returns [(sample1X,sample1Y), (sample2X,sample2Y)...],
#we will have to set rowvar to False. Enough talking, let's code !

def plot_autocorr(N=10000, gibbs_sampler = gibbs_sampler, maxlag=12, rho=0.5):
    samps, lagList = np.array(gibbs_sampler(N,rho)), np.array(range(1,maxlag))
    x1s_cor,x2s_cor = [],[]
    for curLag in lagList:
        #As a reminder, our rows contain observations, and we only have 2 columns for X1 and X2. so samps[lag: removes the first lag observations, which allows to get lag.
        #Also, samps[:-lag] doesn't change the order of the first samples, but reduces the total number to the correct amount to match the first. 
        corr = np.corrcoef(samps[curLag:,:] , samps[:-curLag,:], rowvar = False) #Obs are rows.
        
        #For coordinates : basically, in 0 there's X1t, in 1 there's X2t (corresponding samples are close to each other), in 2 there's X1t-lag and in 3 there's X2t-lag
        x1s_cor.append(corr[0,2]) #correlation between X1t,X1t-lag. 
        x2s_cor.append(corr[1,3]) #correlation between X2t,X2t-lag.
        #Other correlations are not interesting. Maybe the one between X1t and X2t that direcetly depends on rho but the rest is useless.
    #Now we plot :
    plt.close('all')
    plt.plot(lagList, x1s_cor, label='Autocorr of X1 in function of lag')
    plt.plot(lagList, x2s_cor, label='Autocorr of X2 in function of lag')
    plt.grid()
    plt.title('Autocorrelations of X1 and X2 plotted in functino of lag')
    plt.show()
```

5. Using what you observe, propose two arguments to add to your Gibbs sampler function to improve the simulation.
What happens when $\rho \to 1$ ?

Two arguments : the <strong> burn-in </strong> and the <strong> one-every-four samples selection </strong> </br>
The burn-in (which removes the first samples from the total sample list) so that the influence of the starting point (which could be very far from the 'probability center =' (probable zone)). For instance if we wanted to estimate a centered white noise and our starting point was very far from (0,0), it would make sense to wait until we came back around 0 before considering our samples are representative of the distribution. That's the burn-in. We started here at (0,0), which is ok, but we would really want to start somewhere 'random' but 'probable', which implies using that burn-in technique. 
</br>
</br>
Correlation-Reducing selection : Instead of taking every other sample (1 sample every two samples), what we would also do is take one sample every 3 or 4 or 10 samples so that the correlation between them falls. We know that the closer two samples are from each other, the higher they're correlated, and we don't want to keep samples that are to correlated to one another. We proved this with the question before with the autocorrelation graph. So by waiting before selecting samples, we reduce this influence of correlation.
</br>
To explain it in detail : what we do at each step is we move along one coordinate each time (x1, then x2, SAMPLE, x1 again, then x2 again, SAMPLE...). Technically, we have a different sample every change of xi but one can clearly see that the correlation would be extremely high if we selected $(x1_t,x2_t)$ AND $(x1_{t+1},x2_t)$. Now, with the previous graph of autocorr functions, we saw that the correlation was still high between $(x1_t,x2_t)$ and $(x1_{t-lag},x2_{t-lag})$ for $lag \in {1,2,3}$ (it's actually quite OK for 3, but whatever), and that it drops for $lag>=4$. So by only selection one in 4 sample pair, we ensure that the effect of autocorrelation has been reduced as much as we could. 
</br>
See below code which includes some modifications.

```python
def gibbs_sampler_every4(n=1000, rho = 0.7):
    """Gibbs sampler for (X_1, X_2)."""
    X = []
    x1,x2 = 0,0 #Origin of the samples, center of the target distribution (we don't want to start far from it... but we'll get back to that later)
    for i in range(int(n)):
        #we actualize x1 first, by convention, but as the situation is symmetric it doesn't change anything. 
        for j in range(4):
            x1 = np.random.normal(rho*x2, 1-rho**2)
            x2 = np.random.normal(rho*x1, 1-rho**2)
        #This process of going around four times reduces the correlation by augmenting the (hidden) lag between two following samples. Even though they should be as probable.
        #We append the tuple generated from their respective conditional distributions
        X.append([x1,x2])
    return X
```

```python
#Combine both :
def best_gibbs_sampler(n=10000, rho = 0.7, burninRatio = 0.2):
    """Gibbs sampler for (X_1, X_2)."""
    X = []
    x1,x2 = 0,0 #Origin of the samples, center of the target distribution (we don't want to start far from it... but we'll get back to that later)
    for i in range(int(n)):
        #we actualize x1 first, by convention, but as the situation is symmetric it doesn't change anything. 
        x1 = np.random.normal(rho*x2, 1-rho**2)
        x2 = np.random.normal(rho*x1, 1-rho**2)
        for j in range(4):
            x1 = np.random.normal(rho*x2, 1-rho**2)
            x2 = np.random.normal(rho*x1, 1-rho**2)
        #This process of going around four times reduces the correlation by augmenting the (hidden) lag between two following samples. Even though they should be as probable.
        #We append the tuple generated from their respective conditional distributions
        X.append([x1,x2])
    return X[int(n*burninRatio):][:]
```

6. Implement a Random Walk metropolis algorithm to simulate from p using a conditional proposal $\mathcal N(0, \tau I_2)$ and vary $\tau$. 

```python
#Given how the rw_metropolis_hastings signature has been written, we want to define f beforehand : 
def f_12(x, y,rho=rho):
    return np.exp(-(x**2+y**2-2*rho*x*y) / (2*(1-rho**2))) / (2*np.pi*np.sqrt(1-rho**2))
```

```python
def rw_metropolis_hastings(n=10000, rho=rho, f=f_12, bins=100, tau=1., seed=None):
    """Random walk metropolis hastings algorithm for f with Gaussian proposal g."""
    start_t = time.time()
    np.random.seed(seed)
    x1_t,x2_t=0,0 #We have to start somewhere. Let's put our random walker at the very middle of the distribution, so he can go anywhere with the same probability.
    samps = []
    nSamps = 0
    while nSamps<n:
        x1_t_, x2_t_ = np.random.multivariate_normal((x1_t,x2_t),[[tau,0],[0,tau]])
        alpha = f(x1_t_,x2_t_, rho=rho)/f(x1_t,x2_t, rho=rho)
        if alpha>= 1:
            #if the ratio between g(x_) and g(x) is already higher than 1, we necessarily move & update.
            x1_t, x2_t = x1_t_, x2_t_
            
        #if not, we draw a sample from Uniform[0,1] and if it is smaller than our alpha ratio, we keep it.
        uni_ = np.random.uniform(0,1)
        if uni_ <= alpha:
            x1_t, x2_t = x1_t_, x2_t_

        #else: #In case of failure
            #x = x stay in position, not necessary
            #f_x = f_x stay in position, not necessary
        samps.append((x1_t,x2_t))
        nSamps += 1 #wouldn't work with just len(2Darray)...
        
    samps = np.array(samps)
    end_t = time.time()
    #As soon as we have our samples, let's show our result :
    plt.close('all')
    Z_12, absc, ordo = np.histogram2d(samps[:,0],samps[:,1],bins=bins,density=True)
    xgrid, ygrid = np.meshgrid(absc, ordo) 
    plt.contourf(xgrid[1:,1:], ygrid[1:,1:], Z_12, vmin = 0) #For some reason they don't have the same shape, I'm removing the first value to match both at (50,50)
    plt.title('Random walk Metropolis Hastings')
    print('Elapsed time of Metropolis Hastings method : '+str(end_t-start_t)+' seconds')
    return plt.show()
```

```python
rho = 0.5
rw_metropolis_hastings(n=100000, rho=rho, f=f_12, tau=1., seed=None)
```
